```python
import networkx as nx;
```

```python
G = nx.Graph() # initalize an empty graph 
```

```python
# Add nodes per pdf file
```

```python
G.add_node("Jane")
G.add_node("Ike")
G.add_node("Heather")
G.add_node("Garth")
G.add_node("Ed")
G.add_node("Beverly")
G.add_node("Diane")
G.add_node("Andre")
G.add_node("Carol")
G.add_node("Fernando")
G.nodes()
```

```python
# for an undirected graph, the edges are list of tuples so (u,v) means there is an edge from node u to node v
```

```python
# add_edge adds edges from one node to another
G.add_edge("Ike", "Jane")
G.add_edge("Heather", "Ike")
G.add_edge("Heather", "Fernando")
G.add_edge("Heather", "Garth")
G.add_edge("Fernando", "Garth")
G.add_edge("Fernando", "Diane")
G.add_edge("Fernando", "Carol")
G.add_edge("Fernando", "Andre")
G.add_edge("Garth", "Diane")
G.add_edge("Garth", "Ed")
G.add_edge("Garth", "Beverly")
G.add_edge("Ed", "Beverly")
G.add_edge("Ed", "Diane")
G.add_edge("Beverly", "Diane")
G.add_edge("Beverly", "Andre")
G.add_edge("Diane", "Andre")
G.add_edge("Beverly", "Carol")
G.add_edge("Carol", "Diane")
G.add_edge("Carol", "Andre")
G.edges()
```

```python
# let's draw the graph
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(G, with_labels=True) # draw the graph with node labels
```
